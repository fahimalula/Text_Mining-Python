```python
driver.get(data_url)
```

## 4. Do web scraping for multiple same elements automatically (e.g. element called 'description' here) with interval time 2 secs for jump to each elements

```python
import time

posts = driver.find_elements_by_class_name('result-card__full-card-link')
for post in posts:
    post.click()
    driver.find_element_by_class_name("show-more-less-html__button--more").click()
    time.sleep(2)
    desc = driver.find_element_by_class_name  ('description').text
    time.sleep(2)
    print(desc + "----")
    time.sleep(2)
```

## 4. Select element (part of website) you want to scrap (1 element each, manually)

```python
driver.find_element_by_xpath("""//*[@id="main-content"]/div/section/ul/li[2]/a""").click()
```

## Copy and paste texts to any directory in .txt format (p.s. you can use (or run) paperclip library if you want to copy the text)

```python
driver.find_element_by_class_name("show-more-less-html__button--more").click()

posts = driver.find_elements_by_class_name("description")
for post in posts:
    print(post.text)
    file = open(r"C:\Users\LENOVO\Documents\Python Scripts\NLP\dt8\tech_spec3.txt","w",encoding="utf-8")
    file.write(post.text)
    file.close()
    
# import pyperclip
# pyperclip.copy(post.text)
# spam = pyperclip.paste()      
```
